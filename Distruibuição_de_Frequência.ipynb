```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/Companhia_MB.xlsx')
df
```

```python
# frequência: para se ter uma ideia da frequência dos dados. Vou usar a value_count() para fazer essa tarefa.

fabs_estado_civil = df['Estado Civil'].value_counts() # isso será usado para calcular a frequência relativa nas próximas células.
fabs_estado_civil
```

```python
# frequência dos dados na coluna Grau de Instrução. Obviamente, também vou usar a value_count() para fazer isso.

fabs_grau_ins = df['Grau de Instrução'].value_counts()
fabs_grau_ins
```

```python
# agora dá para calcular a frequência relativa, tanto para estado civil, quanto para grau de instrução.

# Como calculo a frquência relativa? pego a frequência (que é a frequência absoluta, que é a quantidade de vezes que a mesma resposta apareceu) dos dados
#obtida através da value_counts()).

# ou seja:

frel_estado_civil = fabs_estado_civil / len(df['Estado Civil'])
frel_estado_civil
```

```python
# frequência relativa de grau de instrução: usando a frequência absoluta de grau de instrução.

frel_grau_instrucao = fabs_grau_ins / len(df['Grau de Instrução'])
frel_grau_instrucao
```

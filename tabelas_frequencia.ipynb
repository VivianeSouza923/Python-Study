```python
# ordenando a coluna idade em ordem crescente para sabem quem é a maior e a menor idade
df['Idade'].sort_values()
```

```python
import math

# b) Categorizar as variáveis quantitativas e criar uma nova variáveis com essas categorias.

# VARIÁVEL Idade

#cálculo do número de classes:

n = 76839
k = 1 + 3.3*math.log10(n)
k
```

```python
amp_idade = df['Idade'].max() - df['Idade'].min()
amp_idade
```

```python
# AMplitude de cada classe:
k = 17

AmplitudeDeClasse = amp_idade / k
AmplitudeDeClasse
```

```python


# definir os limites dos intervalos para categorizzação da variável idade:

bins = [13, 18, 23, 28, 33, 38, 43, 48, 53, 58, 63, 68, 73, 78, 83, 88]

```

```python
# fazendo a coluna do intervalo de classe da idade.
df['Intervalo de Classe - Idade'] = pd.cut(df['Idade'], bins, right=False)
df
```

```python
# VARIÁVEL Renda

#cálculo do número de classes:

n = 76839
k = 1 + 3.3*math.log10(n)
k
```

```python
# ordenando a coluna renda em ordem crescente para sabem quem é a maior e a menor renda
df['Renda'].sort_values()
```

```python
amp_renda = df['Renda'].max() - df['Renda'].min()
amp_renda
```

```python
# AMplitude de cada classe:
k = 17

AmplitudeDeClasse = amp_renda / k
AmplitudeDeClasse
```

```python
# definir os limites dos intervalos para categorizzação da variável renda:

limites_intervalos = [df['Renda'].min() + i * AmplitudeDeClasse for i in range(k + 1)]
limites_intervalos
```

```python
df['Intervalo de Classe - Renda'] = pd.cut(df['Renda'], bins=limites_intervalos, right=False)
df
```

```python
# VARIÁVEL Altura

#cálculo do número de classes:

n = 76839
k = 1 + 3.3*math.log10(n)
k
```

```python
# ordenando a coluna Altura em ordem crescente para sabem quem é a maior e a menor altura
df['Altura'].sort_values()
```

```python
# Remover os pontos de separação de milhares e converter para números inteiros
df['Altura'] = df['Altura'].str.replace('.', '').astype(float).astype(pd.Int64Dtype())
print(df)
```

```python
amp_altura = df['Altura'].max() - df['Altura'].min()
amp_altura
```

```python
# AMplitude de cada classe:
k = 17

AmplitudeDeClasse = amp_altura / k
AmplitudeDeClasse
```

```python
# definir os limites dos intervalos para categorizzação da variável altura:

limites_intervalos = [df['Altura'].min() + i * AmplitudeDeClasse for i in range(k + 1)]
limites_intervalos
```

```python
df['Intervalo de Classe - Altura'] = pd.cut(df['Altura'], bins=limites_intervalos, right=False)
df
```

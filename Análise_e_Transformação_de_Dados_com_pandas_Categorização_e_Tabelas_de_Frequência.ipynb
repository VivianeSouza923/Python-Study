# b) Categorizar as variáveis quantitativas e criar uma nova variáveis com essas categorias.

```python
import math

# para fazer o cálculo de logaritmo da fórmula do número de classes
```

```python
# VARIÁVEL NU_NOTA_CN

# número de observações (n)

n = 16587

k = int(1 + 3.3*math.log10(n))

k
```

```python
# fazer a conversão da NU_NOTA_CN em float, ou seja, vou remover as vírgulas:

df['NU_NOTA_CN'] = df['NU_NOTA_CN'].str.replace(',', '.').astype(float)
df
```

```python
# cálculo da amplitude total de NU_NOTA_CN:

amp_NU_NOTA_CN = df['NU_NOTA_CN'].max() - df['NU_NOTA_CN'].min()
amp_NU_NOTA_CN
```

```python
# CÁLCULO DA AMPLITUDE DE CADA CLASSE:
amp_classe = amp_NU_NOTA_CN / k

amp_classe
```

```python
# definir os limites de intervalo para categorização da variável NU_NOTA_CN

bins = [df['NU_NOTA_CN'].min() + i * amp_classe for i in range(k + 1)]
```

```python
# inserir a coluna na base da categorização da variável NU_NOTA_CN

df['Intervalo de Classe - NU_NOTA_CN'] = pd.cut(df['NU_NOTA_CN'], bins, right=False)
df
```

```python
# NU_NOTA_CH

# número de observações (n)

n = 16587

k = int(1 + 3.3*math.log10(n))

k
```

```python
# conversão dos dados object da coluna para float

df['NU_NOTA_CH'] = df['NU_NOTA_CH'].str.replace(',', '.').astype(float)
df
```

```python
# cálculo da amplitude total de NU_NOTA_CH:

amp_total = df['NU_NOTA_CH'].max() - df['NU_NOTA_CH'].min()
amp_total
```

```python
# cálculo da amplitude de cada classe
# amplitude = amplitude total / k

amplitude_classe = amp_total / k
amplitude_classe
```

```python
# definir os limites de intervalo para categorização:

bins = [df['NU_NOTA_CH'].min() + i * amplitude_classe for i in range(k + 1)]
bins
```

```python
# inserir a coluna da categorização na base
df['Intervalo de Classe - NU_NOTA_CH'] = pd.cut(df['NU_NOTA_CH'], bins, right=False)
df
```

# b) Categorizar as vari√°veis quantitativas e criar uma nova vari√°veis com essas categorias.

```python
import math

# para fazer o c√°lculo de logaritmo da f√≥rmula do n√∫mero de classes
```

```python
# VARI√ÅVEL NU_NOTA_CN

# n√∫mero de observa√ß√µes (n)

n = 16587

k = int(1 + 3.3*math.log10(n))

k
```

```python
# c√°lculo da amplitude total de NU_NOTA_CN:

amp_NU_NOTA_CN = df['NU_NOTA_CN'].max() - df['NU_NOTA_CN'].min()
amp_NU_NOTA_CN
```

```python
# C√ÅLCULO DA AMPLITUDE DE CADA CLASSE:
amp_classe = amp_NU_NOTA_CN / k

amp_classe
```

```python
# definir os limites de intervalo para categoriza√ß√£o da vari√°vel NU_NOTA_CN

bins = [df['NU_NOTA_CN'].min() + i * amp_classe for i in range(k + 1)]
```

```python
# inserir a coluna na base da categoriza√ß√£o da vari√°vel NU_NOTA_CN

df['Intervalo de Classe - NU_NOTA_CN'] = pd.cut(df['NU_NOTA_CN'], bins, right=False)
df
```
